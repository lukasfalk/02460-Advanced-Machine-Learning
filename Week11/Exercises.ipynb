## Exercise A: Graph convolutions

### Q A.1

Signal will just be itself. You basically just multiply the signal with 1

### Q A.2

Convolution is commutative so we look at x[n]:
$$D^1 x[n] = x[n-1], D^2 x[n] = x[n-2], D^3 x[n] = x[n-3]$$
$$D^k x[n] = x[n-k]$$
Thus
$$\sum_{k=0}^{\infty}h[k]D^kx[n] = \sum_{k=0}^{\infty}h[k]x[n-k]$$

### Q A.3

We have that
$$y[n] = \sum_{n=0}^{N}h[n]D^n x = \sum_{n=0}^{N}h[n]A^n x$$
$$y[n] = \sum_{n=0}^{N}h[n]A^n x$$
So it's just the sum of h[n] from 0 to N, times the constant x
